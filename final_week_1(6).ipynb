```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
booking=pd.read_csv(r'C:\Users\AMCT\Downloads\first inten project.csv')
booking['number of people']=booking['number of adults']+booking['number of children']
booking['total of day booking']=booking['number of week nights']+booking['number of weekend nights']
booking['queuing system']=booking['lead time']-booking['total of day booking']
print(booking.columns)
print(booking.head())
print(booking.info())
print(booking.describe())
print(booking.shape)
print(booking.isnull().sum())

booking['date of reservation'] = pd.to_datetime(booking['date of reservation'], errors='coerce')

# Add columns for year and month
booking['year'] = booking['date of reservation'].dt.year
booking['month'] = booking['date of reservation'].dt.month
years = booking['year'].unique()
print(years)

# Group bookings by year, month, and booking status
grouped_booking = booking.groupby(['year', 'month', 'booking status']).size().unstack(fill_value=0)
# Reset index for plotting
grouped_booking = grouped_booking.reset_index()

# Plotting
fig, ax = plt.subplots(figsize=(14, 8))

# Iterate over booking statuses and plot each as a separate bar
for status in grouped_booking.columns[2:]:  # Skip 'year' and 'month' columns
    ax.bar(
        grouped_booking.index, 
        grouped_booking[status], 
        label=status
    )

# Adding labels and title
ax.set_xlabel('Year-Month')
ax.set_ylabel('Number of Bookings')
ax.set_title('Number of Bookings for Each Year and Month by Status')
ax.legend()

# Customize x-ticks to show year and month
ax.set_xticks(grouped_booking.index)
ax.set_xticklabels(grouped_booking.apply(lambda row: f"{int(row['year'])}-{int(row['month'])}", axis=1), rotation=90)

# Display the plot
plt.tight_layout()
plt.show()
booking['year'] = booking['date of reservation'].dt.year
booking['month'] = booking['date of reservation'].dt.month

# Filter bookings for the year 2015
booking_2015 = booking[booking['year'] == 2015]

# Calculate the sum of 'number of adults' and 'number of children' for bookings in 2015
total_adults_2015 = booking_2015['number of adults'].sum()
total_children_2015 = booking_2015['number of children'].sum()

# Print the results
print(f"Total number of adults in bookings for the year 2015: {total_adults_2015}")
print(f"Total number of children in bookings for the year 2015: {total_children_2015}")
booking['year'] = booking['date of reservation'].dt.year
booking['month'] = booking['date of reservation'].dt.month

# Filter bookings for the year 2016
booking_2016 = booking[booking['year'] == 2016]

# Calculate the sum of 'number of adults' and 'number of children' for bookings in 2015
total_adults_2016 = booking_2016['number of adults'].count()
total_children_2016 = booking_2016['number of children'].sum()

# Print the results
print(f"Total number of adults in bookings for the year 2016: {total_adults_2016}")
print(f"Total number of children in bookings for the year 2016: {total_children_2016}")

# Filter bookings for the year 2017
booking_2017 = booking[booking['year'] == 2017]

# Calculate the sum of 'number of adults' and 'number of children' for bookings in 2015
total_adults_2017 = booking_2017['number of adults'].sum()
total_children_2017 = booking_2017['number of children'].sum()

# Print the results
print(f"Total number of adults in bookings for the year 2017: {total_adults_2017}")
print(f"Total number of children in bookings for the year 2017: {total_children_2017}")

# Filter bookings for the year 2015
booking_2018 = booking[booking['year'] == 2018]

# Calculate the sum of 'number of adults' and 'number of children' for bookings in 2018
total_adults_2018 = booking_2018['number of adults'].sum()
total_children_2018 = booking_2018['number of children'].sum()

# Print the results
print(f"Total number of adults in bookings for the year 2018: {total_adults_2018}")
print(f"Total number of children in bookings for the year 2018: {total_children_2018}")
booking_2015 = booking[booking['year'] == 2015]

# Calculate the sum of 'number of week nights' and 'number of weekend nights' for bookings in 2015
total_week_nights_2015 = booking_2015['number of week nights'].sum()
total_weekend_nights_2015 = booking_2015['number of weekend nights'].sum()

# Print the results
print(f"Total number of week nights in bookings for the year 2015: {total_week_nights_2015}")
print(f"Total number of weekend nights in bookings for the year 2015: {total_weekend_nights_2015}")
booking_2015 = booking[booking['year'] == 2015]

# Calculate the sum of 'number of week nights' and 'number of weekend nights' for bookings in 2016
total_week_nights_2016 = booking_2016['number of week nights'].sum()
total_weekend_nights_2016 = booking_2016['number of weekend nights'].sum()

# Print the results
print(f"Total number of week nights in bookings for the year 2016: {total_week_nights_2016}")
print(f"Total number of weekend nights in bookings for the year 2016: {total_weekend_nights_2016}")
booking_2016 = booking[booking['year'] == 2016]

# Calculate the sum of 'number of week nights' and 'number of weekend nights' for bookings in 2017
total_week_nights_2017 = booking_2017['number of week nights'].sum()
total_weekend_nights_2017 = booking_2017['number of weekend nights'].sum()

# Print the results
print(f"Total number of week nights in bookings for the year 2017: {total_week_nights_2017}")
print(f"Total number of weekend nights in bookings for the year 2017: {total_weekend_nights_2017}")
booking_2017 = booking[booking['year'] == 2017]

# Calculate the sum of 'number of week nights' and 'number of weekend nights' for bookings in 2018
total_week_nights_2018 = booking_2018['number of week nights'].sum()
total_weekend_nights_2018 = booking_2018['number of weekend nights'].sum()

# Print the results
print(f"Total number of week nights in bookings for the year 2018: {total_week_nights_2018}")
print(f"Total number of weekend nights in bookings for the year 2018: {total_weekend_nights_2018}")
filtered_booking = booking[(booking['number of week nights'] > 0) & (booking['car parking space'] > 0)]

# Count the number of such bookings
count_bookings_with_parking = filtered_booking.shape[0]

# Print the result
print(f"The number of bookings with car parking space and number of week nights: {count_bookings_with_parking}")
filtered_booking = booking[(booking['number of weekend nights'] > 0) & (booking['car parking space'] > 0)]

# Count the number of such bookings
count_bookings_with_parking = filtered_booking.shape[0]

# Print the result
print(f"The number of bookings with car parking space and  weekend nights: {count_bookings_with_parking}")
print('__________________________________________________')
print(booking[booking['booking status']=='Canceled'])
```

```python
import matplotlib.pyplot as plt

years = [2015, 2016, 2017, 2018]
adults = [1, 1, 11466, 55416]
children = [1, 0, 429, 3387]
plt.figure(figsize=(12, 10))
colors = ['#ff9999', '#66b3ff']

for i, year in enumerate(years):
    plt.subplot(2, 2, i+1)
    plt.pie([children[i], adults[i]], labels=['Children', 'Adults'], colors=colors, autopct='%1.1f%%', startangle=140)
    plt.title(f'Children vs Adults Booking Distribution ({year})')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

#      
data = {
    'year': [2015, 2016, 2017, 2018],
    'number of adults': [1, 1, 11466, 55416],
    'number of children': [1, 0, 429, 3387],
    'number of people': [2, 1, 11895, 58803]  #     
}

#    DataFrame
booking = pd.DataFrame(data)

#    
years = booking['year']
adults = booking['number of adults']
children = booking['number of children']
total_people = booking['number of people']

#  
plt.figure(figsize=(10, 6))

plt.plot(years, adults, marker='o', linestyle='-', color='b', label='Adults')
plt.plot(years, children, marker='s', linestyle='--', color='r', label='Children')
plt.plot(years, total_people, marker='^', linestyle='-', color='g', label='Total People')

plt.title('Booking Statistics Over Years')
plt.xlabel('Year')
plt.ylabel('Number of People')
plt.xticks(years)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
booking=pd.read_csv(r'C:\Users\AMCT\Downloads\first inten project.csv')
booking['total of day booking']=booking['number of week nights']+booking['number of weekend nights']
booking['queuing system']=booking['lead time']-booking['total of day booking']

plt.figure(figsize=(12, 6))

sns.boxplot(data=booking[['total of day booking', 'queuing system']])
sns.stripplot(data=booking[['total of day booking', 'queuing system']], color='red', alpha=0.5, jitter=True)
plt.title('Box and Strip Plot of Booking Days and Waiting Days')
plt.xlabel('Type')
plt.ylabel('Days')

plt.show()
```
